```python
import csv
from functools import reduce
import numpy as np

def get_numero_de_crimes(crime) :
    return crime['numero_de_crimes']

def mostrar(valores, campo) :
  index = 1
  for valor in valores :
    mensagem =  valor['nome'] + " | %.2f " % valor[campo]
    print('%d ' % index  + mensagem)
    index += 1

cidades = []
with open ('junho.csv', newline ='', encoding = "ISO-8859-1") as csvfile :
  spamreader = csv.reader(csvfile, delimiter=';', quotechar='|')

  for linha in spamreader:
    campos = linha
    if linha[7] == ' Roubos' or linha[0] == 'Total RS':
      continue
  
    cidades.append({
      'nome': campos.pop(0),
      'numero_de_crimes': reduce((lambda x, y: float(x) + float(y)), campos)
    })

  cidades.sort (key = get_numero_de_crimes)

  print('MAIOR CRIMINALIDADE POR CIDADE')
  maior_criminalidade = cidades[-10:]
  mostrar(maior_criminalidade[::-1], 'numero_de_crimes')
  
  print('\nMENOR CRIMINALIDADE POR CIDADE')
  menor_criminalidade = cidades[0:10]
  mostrar(menor_criminalidade, 'numero_de_crimes')


def get_populacao(populacao) :
    return populacao['populacao']

populacoes = []
with open ('população.csv', newline ='', encoding = "ISO-8859-1") as csvfile :
  spamreader = csv.reader(csvfile, delimiter=';', quotechar='|')

  print()
  for linha in spamreader:
    if linha[1] == 'POPULAÇÃO ESTIMADA':
      continue

    populacoes.append({
        'nome': linha[0],
        'populacao': int(linha[1])
    })

  print('MAIORES POPULAÇÕES')
  populacoes.sort (key = get_populacao)
  maior_populacao = populacoes[-10:]

  mostrar(maior_populacao[::-1], 'populacao')


def busca_cidade(cidade_nome) :
  for cidade in cidades :
    if cidade['nome'].lower() == cidade_nome.lower() :
      return cidade

def crimes_por_habitantes(populacao) :
  habitantes = busca_cidade(populacao['nome'])
  if habitantes is None :
    populacao['crimes_por_habitantes'] = 0
    return populacao

  populacao['crimes_por_habitantes'] = habitantes['numero_de_crimes'] / (populacao['populacao'] / 100000)
  return populacao

populacao_por_crime = map(crimes_por_habitantes, populacoes)

def get_crimes_por_habitantes(populacao) :
    return populacao['crimes_por_habitantes']

populacao_por_crime = list(populacao_por_crime)
populacao_por_crime.sort (key = get_crimes_por_habitantes)


print('\nMAIOR CRIMINALIDADE POR 100.000 HABITANTES')
maior_populacao_por_crime = np.asarray(populacao_por_crime)[-10:]
mostrar(maior_populacao_por_crime[::-1], 'crimes_por_habitantes')

print('\nMENOR CRIMINALIDADE POR 100.000 HABITANTES')
menor_populacao_por_crime = np.asarray(populacao_por_crime)[0:10]
mostrar(menor_populacao_por_crime, 'crimes_por_habitantes')
```
